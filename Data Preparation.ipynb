# 2. Data Cleaning

## 2.1. Remove Exact Duplicates

```python
df = df.drop_duplicates()
```

```python
df.info()
```

## 2.2. Convert Data Types

```python
df['Gender'] = df['Gender'].astype('category')
df['Workout_Type'] = df['Workout_Type'].astype('category')
df['Experience_Level'] = df['Experience_Level'].astype('category')
```

```python
df.info()
```

```python
df['Experience_Level'] = df['Experience_Level'].cat.set_categories([1, 2, 3], ordered=True) # Set Experience_Level as ordinal category
```

```python
df['Experience_Level'].dtype
```

## 2.3. Verifying BMI Accuracy

```python
bmi_calc = df['Weight (kg)'] / (df['Height (m)'] ** 2)
mask = (np.abs(df['BMI'] - bmi_calc) / bmi_calc > 0.05) # True if BMI deviates by >5% from calculation
df.loc[mask, 'BMI'] = bmi_calc[mask]
print(f"Recomputed BMI for {mask.sum()} records")
```

## 2.4. Handling Outliers

```python
# Using IQR method to determine outliers
Q1 = df['Calories_Burned'].quantile(0.25)
Q3 = df['Calories_Burned'].quantile(0.75)
IQR = Q3 - Q1
lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR
outliers = df[(df['Calories_Burned'] < lower) | (df['Calories_Burned'] > upper)]

print(f"Number of outliers: {len(outliers)}")
```

```python
df = df[(df['Calories_Burned'] >= lower) & (df['Calories_Burned'] <= upper)]
```

```python
df['Calories_Burned'].describe()
```

## 3.3. Correlation Matrix (Numeric Features)

```python
num_cols = ['Age', 'Weight (kg)', 'Height (m)', 'Max_BPM', 'Avg_BPM',
            'Resting_BPM', 'Session_Duration (hours)', 'Calories_Burned',
            'Fat_Percentage', 'Water_Intake (liters)', 'Workout_Frequency (days/week)', 'BMI']

corr = df[num_cols].corr()
plt.figure()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix for Numeric Features')
plt.show()
```

## Findings:
* Age, weight, height, max_BPM, resting_BPM and BMI have very weak correlation with the target -> *Experiment excluding these features in SAS Viya*
* Weight and BMI have a strong correlation of 0.85, indicating multicollinearity issue, but BMI should still be kept as it is an interaction variable that summarizes the body composition in a medically meaningful way -> *Experiment excluding weight in SAS Viya*

## 3.4. Analyse Relationships Between Key Features

```python
sns.pairplot(df, vars=['Calories_Burned', 'Session_Duration (hours)', 'Fat_Percentage', 'Workout_Frequency (days/week)'],
             hue='Experience_Level', corner=True)
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()
```

## Findings:
* There is a clear relationship between session duration and calories burned
* Fat percentage decreases with experience level (those with experience level of 1 and 2 generally have 20-35% fat while those with experience level of 3 generally have 10-20% fat)
* Workout frequency increases with experience level (those with experience level of 1 workout 1-2 times a week, while those with 2 workout 2-3 times a week, and those with 3 workout 3-4 times a week)

```python
plt.figure()
sns.scatterplot(x='Session_Duration (hours)', y='Calories_Burned',
                hue='Experience_Level', data=df)
plt.title('Session Duration vs. Calories Burned')
plt.show()
```

## Findings:
* The relationship appears linear -> *linear modeling in SAS Viya might be more suitable*
* The slope of the calories burned per hour seems steeper with more experience, meaning the effect of session duration on calories burned could be amplified by experience level -> *experiment using interaction effect in SAS Viya*
